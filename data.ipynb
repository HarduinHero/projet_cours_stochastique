# Code pour trier et filter le dataset

## Contingence ligne

```python
# Chargement des données
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

ot_odr_filename = os.path.join(".", "OT_ODR.csv.bz2")
ot_odr_df = pd.read_csv(ot_odr_filename, compression="bz2", sep=";")

equipements_filename = os.path.join(".", 'EQUIPEMENTS.csv')
equipements_df = pd.read_csv(equipements_filename, sep=";")

merged_df = pd.merge(ot_odr_df, equipements_df, on='EQU_ID', how='inner')

contingency_table= pd.crosstab(merged_df["SYSTEM_N1"],merged_df["LIGNE"])

plt.figure(figsize=(10, 6))
sns.heatmap(contingency_table, annot=False,fmt='d', cmap='coolwarm')
plt.title('Matrice de contingence entre LIGNE et SYSTEM_N2')
plt.show()

print(contingency_table)
```
